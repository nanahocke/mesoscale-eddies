```python
_wind_11_=xr.where(corr.corr_o2_stf_dic_stf > 0, 1,np.nan)
_wind_12_=xr.where((mask.CO2outO2out==1) | (mask.CO2inO2in==1),1,np.nan)
wind_1=xr.where((_wind_11_==1) & (_wind_12_==1), 2,np.nan)
#wind_1=solubility_ctrl.where((mask.CO2outO2out==1) | (mask.CO2inO2in==1),2,np.nan)
solubility_ctrl=xr.where((_wind_11_==1) & (_wind_12_!=1),1,np.nan)
```

```python
_wind_21_=xr.where(corr.corr_o2_stf_dic_stf<0, 1,np.nan)
_wind_22_=xr.where((mask.CO2outO2in==1) | (mask.CO2inO2out==1),1,np.nan)
wind_2=xr.where((_wind_21_==1) & (_wind_22_==1), 3,np.nan)
biologie_ctrl=xr.where((corr.corr_o2_stf_dic_stf<0) & (corr.corr_jp_dic_stf_ctrl > 0) & (wind_2 != 3), 4,np.nan)
biologie_ctrl_wind=xr.where((corr.corr_o2_stf_dic_stf<0) & (corr.corr_jp_dic_stf_ctrl > 0) & (wind_2 == 3), 5,np.nan)
upwelling_ctrl=xr.where((corr.corr_o2_stf_dic_stf<0) & (corr.corr_mld_dic_stf_ctrl > 0)& (wind_2 != 3), 6,np.nan)
upwelling_ctrl_wind=xr.where((corr.corr_o2_stf_dic_stf<0) & (corr.corr_mld_dic_stf_ctrl > 0)& (wind_2 == 3), 7,np.nan)
__both1__=xr.where((biologie_ctrl==4) | (biologie_ctrl_wind==5), 1, np.nan)
__both2__=xr.where((upwelling_ctrl==6) | (upwelling_ctrl_wind==7), 1, np.nan)
both_ctrl=xr.where((__both1__==1)& (__both2__==1) & (wind_2!=3), 8,np.nan)
both_ctrl_wind=xr.where(((__both1__==1)& (__both2__==1) ) & (wind_2==3), 9,np.nan)


##Dominanz von Bio oder Upwelling Korrelation mit CO2?
__both_bio=xr.where((both_ctrl==8)& (corr.corr_jp_dic_stf_ctrl >corr.corr_mld_dic_stf_ctrl),1,np.nan)
__just_bio=xr.where((biologie_ctrl==4) & (both_ctrl !=8),1,np.nan)
biologie_ctrl=xr.where((__both_bio==1)| (__just_bio==1), 4, np.nan)

__both_bio_wind=xr.where((both_ctrl_wind==8)& (corr.corr_jp_dic_stf_ctrl >corr.corr_mld_dic_stf_ctrl),1,np.nan)
__just_bio_wind=xr.where((biologie_ctrl_wind==5) & (both_ctrl_wind !=9),1,np.nan)
biologie_ctrl_wind=xr.where((__both_bio_wind==1)| (__just_bio_wind==1), 5, np.nan)


__both_upw=xr.where((both_ctrl==8)& (corr.corr_jp_dic_stf_ctrl <corr.corr_mld_dic_stf_ctrl),1,np.nan)
__just_upw=xr.where((upwelling_ctrl==6) & (both_ctrl !=8),1,np.nan)
upwelling_ctrl=xr.where((__both_upw==1)| (__just_upw==1), 6, np.nan)

__both_upw_wind=xr.where((both_ctrl_wind==9)& (corr.corr_jp_dic_stf_ctrl <corr.corr_mld_dic_stf_ctrl),1,np.nan)
__just_upw_wind=xr.where((upwelling_ctrl_wind==7) & (both_ctrl_wind !=9),1,np.nan)
upwelling_ctrl_wind=xr.where((__both_upw_wind==1)| (__just_upw_wind==1), 7, np.nan)

##dom=mask area where bio or upwelling are dominating in an area which shows both signals
dom = xr.full_like(__both_bio, fill_value=0, dtype=int)
dom=dom.where(__both_bio!=1,1)
dom=dom.where(__both_upw!=1,1)
dom = dom.where(dom!=0, np.nan)

dom_wind = xr.full_like(__both_bio_wind, fill_value=0, dtype=int)
dom_wind=dom_wind.where(__both_bio_wind!=1,1)
dom_wind=dom_wind.where(__both_upw_wind!=1,1)
dom_wind = dom_wind.where(dom_wind!=0, np.nan)

##take out area from initial 'both' which is now dominated by bio or upw
both_ctrl=both_ctrl.where(dom.isnull())
both_ctrl_wind=both_ctrl_wind.where(dom_wind.isnull())
```

```python
###combine everything
combined_mask = xr.full_like(solubility_ctrl, fill_value=0, dtype=int)
combined_mask = combined_mask.where(solubility_ctrl!=1, 1)
combined_mask = combined_mask.where(wind_1!=2, 2)
combined_mask = combined_mask.where(wind_2!=3, 3)
combined_mask = combined_mask.where(biologie_ctrl!=4, 4)
combined_mask = combined_mask.where(biologie_ctrl_wind!=5, 5)
combined_mask = combined_mask.where(upwelling_ctrl!=6, 6)
combined_mask = combined_mask.where(upwelling_ctrl_wind!=7, 7)
combined_mask = combined_mask.where(both_ctrl!=8, 8)
combined_mask = combined_mask.where(both_ctrl_wind!=9, 9)
combined_mask = combined_mask.where(combined_mask!=0, np.nan)
```

```python
def weighted_means(data, names):
    weights = np.cos(np.deg2rad(dsa.dic_stf.yt_ocean))
    weights.name = "weights"
    res=[]
    for i,item in enumerate(data):
        item_w=item.weighted(weights).mean(("xt_ocean", "yt_ocean"))
        item_w.name=names[i]
        res.append(item_w)
    res=xr.merge(res)
    return res
```
