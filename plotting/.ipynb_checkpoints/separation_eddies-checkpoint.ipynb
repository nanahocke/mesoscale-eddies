```python
ds_mom=xr.open_dataset('/gxfs_work/geomar/smomw577/mesoscale_eddies/MOM5_concat/0181-0190/MOM5_monthly_0181-0190_noice_all.nc')
pathp='/gxfs_home/geomar/smomw577/plots/Boxcar/0181-0190/'
pathf='/gxfs_work/geomar/smomw577/mesoscale_eddies/BOX_filtered/0181-0190/'
dsa=xr.open_dataset(pathf+'3x3box_median_anomaly_monthly_0181-0190_all.nc')
corr=xr.open_dataset(pathf+'3x3box_median_corr_monthly_0181-0190_all.nc')
ds=xr.open_dataset('/gxfs_work/geomar/smomw577/mesoscale_eddies/MOM5_concat/0181-0190/mean_std_control_noice.nc', chunks='auto')
ds_mask=xr.open_dataset(pathf+'mask_gases_total.nc')
```

```python
ds_mom=fix_grid(ds_mom)
dsa=fix_grid(dsa)
corr=fix_grid(corr)
ds=fix_grid(ds)
ds_mask=fix_grid(ds_mask)
```

```python
pos=dsa.SST.where(dsa.SST>0)
neg=dsa.SST.where(dsa.SST<0)
```

```python
dic_pos=dsa.dic_stf.where(pos>0)
dic_neg=dsa.dic_stf.where(neg<0)
o2_pos=dsa.o2_stf.where(pos>0)
o2_neg=dsa.o2_stf.where(neg<0)
```

```python
fig, axs= plt.subplots(nrows=2, ncols=2 , figsize=(16,9),subplot_kw={'projection': ccrs.PlateCarree()})
axs=axs.flatten()
dic_pos.mean(dim='time').plot(ax=axs[0], cmap=cmo.balance, vmin=-1, vmax=1, x='geolon_t', y='geolat_t', transform=ccrs.PlateCarree(),cbar_kwargs={'orientation':'horizontal'})
dic_neg.mean(dim='time').plot(ax=axs[1], cmap=cmo.balance, vmin=-1, vmax=1, x='geolon_t', y='geolat_t', transform=ccrs.PlateCarree(),cbar_kwargs={'orientation':'horizontal'})
o2_pos.mean(dim='time').plot(ax=axs[2], cmap=cmo.balance, vmin=-4, vmax=4, x='geolon_t', y='geolat_t', transform=ccrs.PlateCarree(),cbar_kwargs={'orientation':'horizontal'})
o2_neg.mean(dim='time').plot(ax=axs[3], cmap=cmo.balance, vmin=-4, vmax=4, x='geolon_t', y='geolat_t', transform=ccrs.PlateCarree(),cbar_kwargs={'orientation':'horizontal'})

for i in range(4):
    gl = axs[i].gridlines(crs=ccrs.PlateCarree(), draw_labels=True,
                      linewidth=1, color='black', alpha=0.5)
    gl.xlabels_top = False
    gl.ylabels_left = False
    axs[i].coastlines()
axs[0].set_title('Anticyclonic FCO2A', fontweight='bold');
axs[1].set_title('Cyclonic FCO2A', fontweight='bold');
axs[2].set_title('Anticyclonic FO2A', fontweight='bold');
axs[3].set_title('Cyclonic FO2A', fontweight='bold')

plt.savefig(pathp+'eddies_separated_gases.png', dpi=300, bbox_inches='tight');
```

```python
def weighted_means(data, names):
    weights = np.cos(np.deg2rad(dsa.dic_stf.yt_ocean))
    weights.name = "weights"
    res=[]
    for i,item in enumerate(data):
        item_w=item.weighted(weights).mean(("xt_ocean", "yt_ocean"))
        item_w.name=names[i]
        res.append(item_w)
    res=xr.merge(res)
    return res
```

```python
weightedmeans=weighted_means([dic_pos,dic_neg, o2_pos, o2_neg], ["AC_CO2","C_CO2", "AC_O2", "C_O2"])
```

```python
weightedmeans=weightedmeans.mean(dim='time')
```

```python
index = ['FCO2', 'FO2']
values = pd.DataFrame({'Anticyclone': [weightedmeans['AC_CO2'].values, weightedmeans['AC_O2'].values],
                   'Cyclone': [weightedmeans['C_CO2'].values, weightedmeans['C_O2'].values]}, index=index).astype(float)
```

```python
fig,axs=plt.subplots(1,1, figsize=(7,3), sharey=True)

axs.set_title('Global Mean', fontweight='bold')
values.plot.barh(ax=axs, color=['firebrick', 'mediumblue'])
axs.set_xlabel('[mol/m-2 yr-1]')
axs.axvline(0, c='black', linewidth=1)
plt.grid()
plt.savefig(pathp+'eddies_separated_gases_glomean.png', dpi=300, bbox_inches='tight');
```
