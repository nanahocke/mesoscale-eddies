```python
import numpy as np
import dask , dask.distributed
import xarray as xr
import matplotlib.pyplot as plt
import cartopy.crs as ccrs
import cmocean.cm as cmo
import warnings
warnings.simplefilter('ignore')
import dask_jobqueue
from astropy.convolution import Box2DKernel, convolve
from cartopy.mpl.ticker import (LongitudeFormatter, LatitudeFormatter,
                                LatitudeLocator, LongitudeLocator)
import matplotlib.gridspec as gridspec
import cartopy.feature as cfeature
```

```python
def fix_grid(ds):
    grid=xr.open_dataset('/gxfs_work/geomar/smomw577/mesoscale_eddies/MOM5/gridfiles/CM2.6_grid_spec.nc')[['geolat_t', 'geolon_t']]
    grid=grid.rename({'gridlat_t': 'yt_ocean', 'gridlon_t':'xt_ocean'})
    glat=grid.geolat_t.interp_like(ds)
    glon=grid.geolon_t.interp_like(ds)
    ds=ds.assign_coords({'geolat_t':glat, 'geolon_t': glon})
    return ds
```

# plot and filter same sign regions

```python
ds_corr=xr.load_dataset('/gxfs_work/geomar/smomw577/mesoscale_eddies/BOX_filtered/0181-0190/3x3box_corr_monthly_0181-0190.nc')
```

```python
ds_corr=fix_grid(ds_corr)
```

```python
pathp='/gxfs_home/geomar/smomw577/plots/Boxcar/0181-0190/'
```

```python
def samesign(ds_corr):
    ###start with same sign and SSH: O2, CO2
    SSH_O2_DIC_p=ds_corr.corr_ssh_o2_stf.where(ds_corr.corr_ssh_o2_stf >0).where(ds_corr.corr_ssh_dic_stf >0)
    SSH_O2_DIC_n=ds_corr.corr_ssh_o2_stf.where(ds_corr.corr_ssh_o2_stf <0).where(ds_corr.corr_ssh_dic_stf <0)
    SSH_O2_DIC=SSH_O2_DIC_p.combine_first(SSH_O2_DIC_n)
    
    ###start with same sign and SSH: O2, SST
    SSH_SST_O2_p=ds_corr.corr_ssh_o2_stf.where(ds_corr.corr_ssh_o2_stf >0).where(ds_corr.corr_ssh_SST >0)
    SSH_SST_O2_n=ds_corr.corr_ssh_o2_stf.where(ds_corr.corr_ssh_o2_stf <0).where(ds_corr.corr_ssh_SST <0)
    SSH_SST_O2=SSH_SST_O2_p.combine_first(SSH_SST_O2_n)
    
    ###start with same sign and SSH: CO2, SST
    SSH_SST_DIC_p=ds_corr.corr_ssh_dic_stf.where(ds_corr.corr_ssh_dic_stf >0).where(ds_corr.corr_ssh_SST >0)
    SSH_SST_DIC_n=ds_corr.corr_ssh_dic_stf.where(ds_corr.corr_ssh_dic_stf <0).where(ds_corr.corr_ssh_SST <0)
    SSH_SST_DIC=SSH_SST_DIC_p.combine_first(SSH_SST_DIC_n)
    
    ###start with same sign and SST: O2, CO2
    SST_O2_DIC_p=ds_corr.corr_sst_o2_stf.where(ds_corr.corr_sst_o2_stf >0).where(ds_corr.corr_sst_dic_stf >0)
    SST_O2_DIC_n=ds_corr.corr_sst_o2_stf.where(ds_corr.corr_sst_o2_stf <0).where(ds_corr.corr_sst_dic_stf <0)
    SST_O2_DIC=SST_O2_DIC_p.combine_first(SST_O2_DIC_n)

    ###plotting
    fig, axs= plt.subplots(nrows=2, ncols=2 , figsize=(12,7),subplot_kw={'projection': ccrs.PlateCarree()})
    axs=axs.flatten()
    
    p=SSH_O2_DIC.plot(ax=axs[0], x='geolon_t', y='geolat_t', vmin=-1, add_colorbar=False, vmax=1,transform = ccrs.PlateCarree(), cmap='coolwarm',levels=np.linspace(-1,1,3))
    p.axes.coastlines()
    axs[0].set_title('Regions of same sign SSH-correlation:\nO2 & CO2')
    
    p=SSH_SST_O2.plot(ax=axs[1], x='geolon_t', y='geolat_t', vmin=-1, vmax=1, add_colorbar=False,transform = ccrs.PlateCarree(), cmap='coolwarm',levels=np.linspace(-1,1,3))
    p.axes.coastlines()
    axs[1].set_title('Regions of same sign SSH-correlation:\nO2 & SST');
    
    p=SSH_SST_DIC.plot(ax=axs[2], x='geolon_t', y='geolat_t', vmin=-1, vmax=1, add_colorbar=False,transform = ccrs.PlateCarree(), cmap='coolwarm',levels=np.linspace(-1,1,3))
    p.axes.coastlines()
    axs[2].set_title('Regions of same sign SSH-correlation:\nCO2 & SST');
    
    p=SST_O2_DIC.plot(ax=axs[3], x='geolon_t', y='geolat_t', vmin=-1, vmax=1, add_colorbar=False,transform = ccrs.PlateCarree(), cmap='coolwarm',levels=np.linspace(-1,1,3))
    p.axes.coastlines()
    axs[3].set_title('Regions of same sign SST-correlation:\nO2 & CO2');
    
    for i in range(0,4):
        gl = axs[i].gridlines(crs=ccrs.PlateCarree(), draw_labels=True,
                                  linewidth=1, color='black', alpha=0)
        gl.xlabels_top = False
        gl.ylabels_left = False
   
    plt.tight_layout()
    plt.savefig(pathp+'regional_corr_0181-0190.png', bbox_inches="tight", dpi=300);
    return SSH_O2_DIC, SSH_SST_O2, SSH_SST_DIC, SST_O2_DIC
```

```python
SSH_O2_DIC, SSH_SST_O2, SSH_SST_DIC, SST_O2_DIC=samesign(ds_corr)
```

# coarser grid

```python
def coarsen(ds, var_list, name_list):
    grid=xr.zeros_like(ds.corr_ssh_sst)
    grid=grid[::10,::10]
    for i, item in enumerate(var_list):
        item.name= name_list[i]
    ds=xr.merge(var_list)
    ds_coarse=[]
    for item in list(ds.keys()):
        coarse=ds[item].interp_like(grid)
        ds_coarse.append(coarse)
    ds_coarse=xr.merge(ds_coarse)  
    return ds_coarse
```

```python
var_list=[SSH_O2_DIC, SSH_SST_O2, SSH_SST_DIC, SST_O2_DIC]
name_list=['SSH_O2_DIC', 'SSH_SST_O2','SSH_SST_DIC','SST_O2_DIC']
```

```python
ds_coarse=coarsen(ds_corr, var_list, name_list)
```

```python
###plotting
fig, axs= plt.subplots(nrows=2, ncols=2 , figsize=(12,7),subplot_kw={'projection': ccrs.PlateCarree()})
axs=axs.flatten()

p=ds_coarse.SSH_O2_DIC.plot.contourf(ax=axs[0], vmin=-1, add_colorbar=False, vmax=1,transform = ccrs.PlateCarree(), cmap='coolwarm',levels=np.linspace(-1,1,3))
p.axes.coastlines()
axs[0].set_title('Regions of same sign SSH-correlation:\nO2 & CO2')

p=ds_coarse.SSH_SST_O2.plot.contourf(ax=axs[1], vmin=-1, vmax=1, add_colorbar=False,transform = ccrs.PlateCarree(), cmap='coolwarm',levels=np.linspace(-1,1,3))
p.axes.coastlines()
axs[1].set_title('Regions of same sign SSH-correlation:\nO2 & SST');

p=ds_coarse.SSH_SST_DIC.plot.contourf(ax=axs[2], vmin=-1, vmax=1, add_colorbar=False,transform = ccrs.PlateCarree(), cmap='coolwarm',levels=np.linspace(-1,1,3))
p.axes.coastlines()
axs[2].set_title('Regions of same sign SSH-correlation:\nCO2 & SST');

p=ds_coarse.SST_O2_DIC.plot.contourf(ax=axs[3], vmin=-1, vmax=1, add_colorbar=False,transform = ccrs.PlateCarree(), cmap='coolwarm',levels=np.linspace(-1,1,3))
p.axes.coastlines()
axs[3].set_title('Regions of same sign SST-correlation:\nO2 & CO2');

for i in range(0,4):
    gl = axs[i].gridlines(crs=ccrs.PlateCarree(), draw_labels=True,
                              linewidth=1, color='black', alpha=0)
    gl.xlabels_top = False
    gl.ylabels_left = False

plt.tight_layout()
plt.savefig('/gxfs_home/geomar/smomw577/plots/Boxcar/correlations/corr_regions/regional_corr_coarse.png', bbox_inches="tight", dpi=300);
```
